```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
irysy = pd.read_csv("dane/iris.data", names=["sep length", "sep width", "pet length", "pet width", "class"])
```

### Rozkład atrybutów

```python
print("Średnie arytmetyczne: \n")
print(irysy.mean())
```

```python
print("Odchylenia standardowe: \n")
print(irysy.std())
```

```python
print("Mediany:\n")
print(irysy.median())
```

```python
print("Wartości minimalne:\n")
print(irysy.min())
```

```python
print("Wartości maksymalne:\n")
print(irysy.max())
```

```python
print("Liczebność danych jakościowych:\n")
print(irysy.groupby("class").size())
```

```python
print("Częstość danych jakościowych:\n")
licz = irysy.groupby("class").size()
print(licz*100/licz.sum())
```

### Analiza korelacji atrybutów

```python
print("Macierz korelacji liniowej:\n")
print(irysy.corr())
```

```python
sns.distplot(irysy["pet length"], bins=20, kde=False).set_title("Histogram dla długości płatka")
```

```python
sns.distplot(irysy["pet width"], bins=20, kde=False).set_title("Histogram dla szerokości płatka")
```

```python
korelacje = irysy.corr()
sns.heatmap(korelacje, annot = True, annot_kws = {"size": 9}, vmin = -1, vmax = 1, center = 0).set_title("Mapa cieplna dla najbardziej skorelowanych atrybutów")
```

```python
pd.plotting.scatter_matrix(irysy, figsize = (15, 15), s = 800)
plt.suptitle('Rozkład poszczególnych par atrybutów')
```

```python
sns.regplot(x = "pet length", y = "pet width", data = irysy,scatter_kws={"color": "black"} ,line_kws={"color": "red"}).set_title("Krzywa regresji liniowej")
```
